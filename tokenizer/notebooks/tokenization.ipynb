Lets implement Byte Pair Encoding ALgorithm, which is nothing but, you search the most occuring pair of bytes and replacing them with single byte and doing this recursively

Lets encode them and turn them into a list first

```python
tokens = text.encode('utf-8')
tokens = list(map(int, tokens))
print(text)
print('------------')
print("length: " , len(text))
print('------------')
print(tokens)
print("length: ", len(tokens))
```

Finding the most frequently occured pair

```python
def find_pair(tokens):
    f = {}
    for pair in zip(tokens, tokens[1:]):
        f[pair] = f.get(pair, 0) + 1
    
    max_pair = max(f, key=f.get)
    return max_pair
```

```python
def merge(tokens, max_pair, new_value):
    new_tokens = []

    i = 0
    while i < len(tokens):
        if i < len(tokens) - 1 and tokens[i] == max_pair[0] and tokens[i+1] == max_pair[1]:
            new_tokens.append(256)
            i += 2
        else:
            new_tokens.append(tokens[i])
            i += 1

    return new_tokens
```
